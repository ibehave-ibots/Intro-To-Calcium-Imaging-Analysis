## `numpy` for 2D arrays (image)

Images can be represented by 2-dimensional `numpy` array represented by rows and columns. If an image has 2x2 pixels where the bottom left pixel has a value of 1 and the rest have 0, we can represent this in numpy array as 

[

        [0, 0], # row1

        [1, 0]  # row2
    
]

**Example** Create a 2x2 frame where the bottom left pixel has value of 1 and the rest are 0. What are its dimensions, shape and dtype?

```python
frame = np.array(
    [
        [0, 0],
        [1, 0]
    ]
)
frame.ndim, frame.shape, frame.dtype
```

**Example** For a random image of 256x256 with pixel values between 0 and 255, set any pixel with value less than 250 to zero

```python
image = np.random.randint(0, 256, (256, 256))
image_filter = image.copy()
image_filter[image_filter < 250] = 0

plt.subplot(1, 2, 1)
plt.imshow(image)

plt.subplot(1, 2, 2)
plt.imshow(image_filter)
```

Read `img/1p_maxproj.png` as numpy array and set any pixel with value less than the mean value to 0

```python
image = plt.imread('img/1p_maxproj.png')
image_filter = image.copy()
filter_value = np.mean(image_filter)
image_filter[image_filter < filter_value] = 0

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')

plt.subplot(1, 2, 2)
plt.imshow(image_filter, cmap='gray')
```

Read `img/1p_maxproj.png` as numpy array and set any pixel with value less than 50 times the minimum value to the mean value

```python
image = plt.imread('img/1p_maxproj.png')
image_filter = image.copy()
filter_value = np.min(image_filter) * 50.
image_filter[image_filter < filter_value] = np.mean(image)

plt.subplot(1, 2, 1)
plt.imshow(image)

plt.subplot(1, 2, 2)
plt.imshow(image_filter)
```

Read `img/1p_maxproj.png` as numpy array and set any pixel with value less than (mean - 3*std) to zero

```python
image = plt.imread('img/1p_maxproj.png')
image_filter = image.copy()
filter_value = np.mean(image_filter) - 3.0*np.std(image_filter)
image_filter[image_filter < filter_value] = 0.

plt.subplot(1, 2, 1)
plt.imshow(image)

plt.subplot(1, 2, 2)
plt.imshow(image_filter)
```

Read `img/1p_maxproj.png` as numpy array and set any pixel with value less than (mean + 3*std) to zero

```python
image = plt.imread('img/1p_maxproj.png')
image_filter = image.copy()
filter_value = np.mean(image_filter) + 3.0*np.std(image_filter)
image_filter[image_filter < filter_value] = 0.

plt.subplot(1, 2, 1)
plt.imshow(image)

plt.subplot(1, 2, 2)
plt.imshow(image_filter)
```

## Exploring `cmap`, `vmin`, and `vmax` to see details of an image

**Example** Generate a 2D matrix representing a sinusoidal wave pattern using NumPy and visualize it with a grayscale colormap.

```python
sin_wave = np.sin(np.linspace(-np.pi * 4, np.pi * 4, 200))
sin_matrix = np.tile(sin_wave, (200, 1))
plt.imshow(sin_matrix, cmap='gray');
```

```python
activity_data = np.random.randint(0, 256, (256, 256))
```

Create an exercise to visualize the bottom 25% of activity levels in a dataset, using the 75th percentile as the maximum value in the color scale.

```python
significant_activity_vmin = np.percentile(activity_data, 75)  # 75th percentile for top 25%

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.title("Original Activity Data")
plt.imshow(activity_data, cmap='viridis')
plt.colorbar()

plt.subplot(1, 2, 2)
plt.title("Highlighting Bottom 25% Activity")
plt.imshow(activity_data, cmap='viridis', vmin=0, vmax=significant_activity_vmin)
plt.colorbar()
```

Create an exercise to visualize and differentiate the top 25% of activity levels in a dataset, using the 75th percentile as the minimum value in the color scale.

```python
significant_activity_vmin = np.percentile(activity_data, 75)  # 75th percentile for top 25%

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.title("Original Activity Data")
plt.imshow(activity_data, cmap='viridis')
plt.colorbar()

plt.subplot(1, 2, 2)
plt.title("Highlighting Top 25% Activity")
plt.imshow(activity_data, cmap='viridis', vmin=significant_activity_vmin, vmax=255)
plt.colorbar()
```

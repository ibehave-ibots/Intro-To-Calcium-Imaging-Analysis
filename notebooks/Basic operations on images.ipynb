```python
import numpy as np
import matplotlib.pyplot as plt
import tifffile as tiff
```

```python
# Reading frames

with tiff.TiffFile('../data/example_frames_registration.tif') as tif:
    frames = [page.asarray() for page in tif.pages]
```

```python
# Plot frame 0

plt.imshow(frames[0])
```

```python
# Plot frame 0 in grayscale

plt.imshow(frames[0], cmap='gray')
```

```python
# Plot frame 0 in grayscale with colorbar

plt.imshow(frames[0], cmap='gray')
plt.colorbar()
```

```python
# Plot average of three frames

frames_stacked = np.stack(frames)
frame_avg = np.mean(frames_stacked, axis=0)
plt.imshow(frame_avg, cmap='gray')
plt.colorbar()
```

```python
# Plot median of three frames

frames_stacked = np.stack(frames)
frame_median = np.median(frames_stacked, axis=0)
plt.imshow(frame_median, cmap='gray')
plt.colorbar()
```

```python
# Subtract frame 1 from median frame

frames_stacked = np.stack(frames)
frame_median = np.median(frames_stacked, axis=0)
frame_med_sub_1 = frame_median-frames[0]
plt.imshow(frame_med_sub_1, cmap='gray')
plt.colorbar()
```

```python
# Subtract frame 2 from median frame

frames_stacked = np.stack(frames)
frame_median = np.median(frames_stacked, axis=0)
frame_med_sub_2 = frame_median-frames[1]
plt.imshow(frame_med_sub_2, cmap='gray')
plt.colorbar()
```

```python
# Subtract frame 3 from median frame

frames_stacked = np.stack(frames)
frame_median = np.median(frames_stacked, axis=0)
frame_med_sub_3 = frame_median-frames[2]
plt.imshow(frame_med_sub_3, cmap='gray')
plt.colorbar()
```
